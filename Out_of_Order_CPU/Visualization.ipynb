# An example of visualization (of values recorded)

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Data
d1_values = [4, 16, 64, 256]
d2_values = [4, 16, 64, 256]
values = [
    [20572621, 20572621, 20572621, 20572621],
    [6182079, 4475745, 4475742, 4475742],
    [6181852, 822738, 1039037, 1039037],
    [6181852, 821802, 1038917, 1038917]
]

# Preparing data for 3D plot
X, Y = np.meshgrid(d1_values, d2_values)
Z = np.array(values).T  # Transpose to align with X, Y meshgrid

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

# Labels and title
ax.set_xlabel('ROB Entries')
ax.set_ylabel('PhysIntRegs')
ax.set_zlabel('Values')
ax.set_title('ROB Entries vs PhysInt Regs: Number of cycles rename is blocking')

plt.show()
```
